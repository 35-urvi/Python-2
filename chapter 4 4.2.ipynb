```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('car data.csv')
df
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Year']=2025-df['Year']
df
```

```python
df['Seller_Type'].unique()
```

```python
# to conver catagorial into numerical
pd.get_dummies(df)
```

```python
# to remove first colum on same data eg. from cng,diesel,petrol remove cng just show diesal,petrol
pd.get_dummies(df,drop_first=True)
```

```python
df.drop(columns=['Car_Name'],inplace=True)
```

```python
df
```

```python
df=pd.get_dummies(df,drop_first=True)
df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
x=df[['Year','Present_Price','Kms_Driven','Owner','Fuel_Type_Diesel','Fuel_Type_Petrol','Seller_Type_Individual','Transmission_Manual']]
y=df['Selling_Price']  #target column,dependent variable
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=32) #
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
# print(y_pred)
print(model.coef_)
print(model.intercept_)
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
y_pred1=model.predict([[10,3.5,33000,0,0,1,0,1]])
print(y_pred1)
```

```python
df=pd.read_csv('auto-mpg.csv')
df.drop(columns=['car name'],inplace=True)
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df=df[df['horsepower']!='?']
```

```python
df
```

```python
df['horsepower']=df['horsepower'].astype(int)
```

```python
df.info()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
x=df[['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin']]
y=df['horsepower']#target column,dependent variable
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=32) #
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
# print(y_pred)
print(model.coef_)
print(model.intercept_)
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```
