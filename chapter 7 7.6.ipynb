```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score

with open('TrumpLies.html','r',encoding='utf-8') as file:
    html_content=file.read()
soup=BeautifulSoup(html_content)
# print(soup.prettify())
# box=soup.findAll('a',class_='CGtC98')
# # print(box)
box=soup.find('head')
print('title:',box.find('title').get_text(strip=True))

b1=soup.find('h1',class_='custom-hed story-heading interactive-headline').text
title=''
for i in b1:
    title=title+i.strip()
print(title) 

b2=soup.find('footer',class_='m-statement__footer').text
date=b2.split('😊')[1].strip()
print(date)

b3=soup.find('p',class_='secret-word').text
sw=b3
print(sw)

b4=soup.findAll('h1')
h1=[]
for i in b4:
    h1.append(i.get_text(strip=True))
print(h1)
b5=soup.findAll('h2')
h2=[]
for i in b5:
    h2.append(i.get_text(strip=True))
print(h2)

b6=soup.findAll('a',href=True)
links=[]
for link in b6: 
    if (link['href'].startswith('http') or link['href'].startswith('https')):
        links.append(link['href'])
# print(links)

b7=soup.findAll('span',class_='short-desc')
print(len(b7))

date=[]
statement=[]
text=[]
for i in b7:
    d=i.find('strong').get_text(strip=True)
    s=i.get_text(strip=True).split('“')[1].split('”')[0].replace('"','')
    t=i.find('span',class_='short-truth').get_text(strip=True)[1:].replace(')','')
    date.append(d)
    statement.append(s)
    text.append(t)
#     print('Date:',d)
#     print('statement:',s)
#     print('Text:',t)
#     print('--------------')

d={'Date':date,'statement':statement,'text':text}
df=pd.DataFrame(d)
df
```
