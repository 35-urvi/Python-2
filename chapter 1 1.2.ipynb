```python
df.dropna()  # to drop empty data
```

```python
df.dropna(axis=0)   #in axis 0 drop row having empty data
#by default axis 0
```

```python
df.dropna(axis=1)     #in axis 1 drop column having empty data
```

```python
df.dropna(axis=0,subset=['FSD'])   #delete row having empty data in fsd column
```

```python
df.dropna(axis=0,thresh=3)   #remove row where non null value is less than thresh
```

```python
df.dropna(axis=0,how='all')    #how all remove row having all null value
#by default how is any any remove all row having nan value
```

```python
df.fillna(25)    #replace nan with given value
```

```python
df.fillna(df.mean())   # replace nan with mean of data
```

```python
df.fillna(method='bfill')
```

```python
df=pd.DataFrame([[np.nan,2,3,4],[2,3,np.nan,6],[5,6,7,np.nan]])
print(df.fillna(method='bfill'))    #nan value replace with below value
print()
print(df.fillna(method='ffill'))    #nan value replace with upper value
```
