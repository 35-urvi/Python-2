```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data={'year':[2021,2022,2023,2024,2025],
     'sales':[100,120,150,170,200],
     'expenses':[70,80,90,100,110]}
df=pd.DataFrame(data)
```

```python
df
```

```python
plt.stackplot(df['year'],df['sales'],df['expenses'],labels=['sale','expense'])
plt.legend()
```

```python
plt.fill_between(df['year'],df['sales'],df['expenses'])
plt.show()
```

```python
data={'year':[2021,2022,2023,2024,2025],
     'sales':[100,120,150,170,200],
     'expenses':[70,80,170,100,250]}
df=pd.DataFrame(data)
```

```python
df
```

```python
plt.fill_between(df['year'],df['sales'],df['expenses'])
plt.show()
```

```python
plt.fill_between(df['year'],df['sales'],df['expenses'])
plt.plot(df['year'],df['sales'],color='red')
plt.plot(df['year'],df['expenses'],color='green')
plt.show()
```

```python
fig,ax=plt.subplots()
plt.fill_between(df['year'],df['sales'],df['expenses'])
plt.plot(df['year'],df['sales'],color='red')
plt.plot(df['year'],df['expenses'],color='green')
ax.set_xticks([2021,2022,2023,2024,2025])
plt.show()
```

```python
fig,ax=plt.subplots()
ax.set_xticks([2021,2022,2023,2024,2025])
plt.fill_between(df.year,df.sales,df.expenses,
                where=df.sales>df.expenses,
                color='blue',interpolate=True)
plt.plot(df['year'],df['sales'],color='red')
plt.plot(df['year'],df['expenses'],color='green')
plt.show()
```
