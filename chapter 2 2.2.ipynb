```python
url='https://www.gutenberg.org/files/11/11-0.txt'
response=requests.get(url)
text=response.text
```

```python
print(text)
```

```python
alice_wc=WordCloud(background_color='white',max_words=2000,stopwords=stopwords)
alice_wc.generate(text)
plt.imshow(alice_wc)
plt.axis('off')
plt.show()
```

```python
G=nx.Graph()
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_edge(1,2)
G.add_edge(2,3)
G.add_edge(3,1)
nx.draw(G,with_labels=True)
plt.show()
```

```python
G=nx.DiGraph()
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_edge(1,2)
G.add_edge(2,3)
G.add_edge(3,1)
nx.draw(G,with_labels=True)
plt.show()
```

```python
G=nx.DiGraph()
G.add_node(1)
G.add_nodes_from([2,3])
G.add_nodes_from(range(4,7))
G.add_edge(1,2)
G.add_edge(1,1)
G.add_edges_from([(2,3),(3,6),(4,6),(5,6)])
nx.draw(G,with_labels=True,
       node_size=850,
       node_color='green',
       width=5,
       edge_color='red')
plt.show()
```

```python
instagram={'p1':[0,1,1,0,1],
          'P2':[0,0,1,0,1],
          'p3':[1,1,0,1,1],
          'p4':[1,1,1,0,0],
          'p5':[1,1,0,0,0]}
G=nx.DiGraph()
G.add_nodes_from(instagram.keys())
for i,fr in enumerate(instagram):
    for j,fs in enumerate(instagram[fr]):
        if fs:
            G.add_edge(fr,list(instagram.keys())[j])
nx.draw(G,with_labels=True)
```
