```python
changed_name={'Delhi Daredevils':'Delhi Capitals','Kings XI punjab':'Punjab Kings','Rising Pune Supergiants':'Rising Pune Supergiant'}
df=df.replace(changed_name.keys(),changed_name.values())
print(df)
```

```python
df=df.dropna(subset=['WinningTeam'])
```

```python
#To find win %,home win %,away win % and consider Team 1 as home team
teams=df.Team1.unique()
new_df=pd.DataFrame()
data=[]
for team in teams:
    played=df[(df.Team1==team) | (df.Team2==team)].shape[0]
    win=df[df.WinningTeam==team].shape[0]
    win_per=(win/played)*100
    home_win=df[(df.Team1==team) & (df.WinningTeam==team)].shape[0]/df[df["Team1"]==team].shape[0]*100
    away_win=df[(df.Team2==team) & (df.WinningTeam==team)].shape[0]/df[df["Team2"]==team].shape[0]*100
    data.append([team,played,win_per,home_win,away_win])
new_df[['Team','MatchesPlayed','Win%','HomeWin%','AwayWin%']]=data
print(new_df)
```

```python
#to display team name ,played match,win match and points
teams=df.Team1.unique()
new_df=pd.DataFrame()
data=[]
for team in teams:
    played=df[(df.Season=='2022') & ((df.Team1==team) | (df.Team2==team))].shape[0]
    win=df[(df.Season=='2022') & (df.WinningTeam==team)].shape[0]
    point=win*2
    data.append([team,played,win,point])
new_df[['Team','MatchesPlayed','MatchesWon','point']]=data
print(new_df)
```
