```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score

website='https://github.com/topics'
result=requests.get(website).text
soup=BeautifulSoup(result,'html.parser')
# print(soup.prettify())
box=soup.findAll('div',class_='py-4 border-bottom d-flex flex-justify-between')
# print(box)
title=[]
discription=[]
url=[]
for i in box:
    title.append(i.find('p',class_='f3 lh-condensed mb-0 mt-1 Link--primary').get_text())
    discription.append(i.find('p',class_='f5 color-fg-muted mb-0 mt-1').get_text().strip())
    url.append('https://github.com'+(i.find('a',class_='no-underline flex-1 d-flex flex-column'))['href'])
# print(discription)
d={'title':title,'discription':discription,'url':url}
df=pd.DataFrame(d)
df.to_csv('topics.csv',index=None)
df
```

```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score

with open('indeed jobs.html','r',encoding='utf-8') as file:
    html_content=file.read()
soup=BeautifulSoup(html_content)
# print(soup.prettify())
box=soup.findAll('td',class_='resultContent css-1o6lhys eu4oa1w0')
# print(box)
job_role=[]
company=[]
city=[]
pay=[]
for i in box:
    job_role.append(i.find('div',class_='css-pt3vth e37uo190').get_text())
    company.append(i.find('div',class_='css-1afmp4o e37uo190').get_text())
    city.append(i.find('div',class_='css-1restlb eu4oa1w0').text)
    pay.append(i.find('div',class_='jobMetaDataGroup css-qspwa8 eu4oa1w0').text)
# print(pay)
d={'job_role':job_role,'company':company,'city':city,'pay':pay}
df=pd.DataFrame(d)
df
```
